### Nettoyage des données

1) Exclusion des données inexploitables (nutriments non renseignés)

2) Exclusion des données supérieure à la valeur maximum théorique

3) Suppression des valeurs aberrantes des variables quantitatives (méthode de l'interquartille)

4) Exclusion des colonnes comportant trop de valeurs manquantes (-50% des données remplies)

5) Exclusion des doublons

```python
# sampling rows containing not null nutrition data
for col in nutrition:
    data = data[data[col].notnull()]
    
# eliminating values that are impossible by definition
for var in maxvalue.keys():
    data = data[data[var] <= maxvalue[var]]

# eliminating aberrant values in parent dataframe with interquartilles
for var in maxvalue.keys():
    data = data[np.abs(data[var]-data[var].mean())<=(3*data[var].std())]

# droping columns filled less than half
entries = data.shape[0]
for col in data.columns: 
    missing_values = data[col].isnull().sum()
    if missing_values > entries / 2 and col != 'categories' and col != 'created_datetime':
        data = data.drop([col],axis=1)
        
data.drop_duplicates(inplace=True)
```

### ACP

Analyse en Composantes Principales sur les taux de nutriments par portion de 100g.

Objectifs:

  - Trouver des modes communs entre les variables


  - Découvrir les éventuelles variables redondantes.

```python
def make_acp(list_of_cols):
    # choix du nombre de composantes à calculer
    n_comp = len(list_of_cols)
    print(f"Using {n_comp} composants")
    
    # selection des colonnes à prendre en compte dans l'ACP
    data_pca = data[list_of_cols]
    X = data_pca.values
    features = data_pca.columns
    
    # affichage de la matrice de corrélation
    d = pd.DataFrame(data=data, columns=list_of_cols)
    corr = d.corr()
    mask = np.triu(np.ones_like(corr, dtype=bool))
    f, ax = plt.subplots(figsize=(11, 9))
    cmap = sns.diverging_palette(230, 20, as_cmap=True)
    sns.heatmap(corr, mask=mask, cmap=cmap, vmax=.8, center=0,
    square=True, linewidths=.5, cbar_kws={"shrink": .5})
    plt.show()
    
    # Centrage et Réduction
    std_scale = preprocessing.StandardScaler().fit(X)
    X_scaled = std_scale.transform(X)

    # Calcul des composantes principales
    pca = decomposition.PCA(n_components=n_comp)
    pca.fit(X_scaled)

    # Eboulis des valeurs propres
    display_scree_plot(pca)

    # Cercle des corrélations
    pcs = pca.components_
    display_circles(pcs, n_comp, pca, [(0,1)],
                    labels = np.array(features))
    
    plt.show()
```

### ANOVA

Visualisation des données de l'ANOVA sous la forme de boxplot

```python
data[nutrition].plot(kind='box', subplots=True, title="Visualisation des données pour l'ANOVA", figsize=(20,20), layout=(6,4), showfliers=False)
plt.show()
```

Réalisation d'une analyse de variance entre les taux de nutriments par portion de 100g et le nutrigrade.

Après vérification de la normalité des données avec le test de Kolmogorov-Smirnov, on vérifie la contribution des nutriments au nutrigrade.

Hypothèse nulle h0: Il n'y a pas de variation du taux moyen du nutriment X entre les différents nutrigrades.

Hypothèse alternative: Il y a une variation du taux moyen du nutriment X entre les différents nutrigrades.

```python
def eta_squared(x,y):
    moyenne_y = y.mean()
    classes = []
    for classe in x.unique():
        yi_classe = y[x==classe]
        classes.append({'ni': len(yi_classe),
                        'moyenne_classe': yi_classe.mean()})
    SCT = sum([(yj-moyenne_y)**2 for yj in y])
    SCE = sum([c['ni']*(c['moyenne_classe']-moyenne_y)**2 for c in classes])

    return SCE/SCT

def nutrianova(working_df, nutriment): 
    X = "nutrition_grade_fr" # qualitative
    Y = nutriment # quantitative
    
    sous_echantillon = working_df[working_df[X].notnull()]
    sous_echantillon = sous_echantillon[sous_echantillon[Y].notnull()]
    
    es = round(eta_squared(sous_echantillon[X],sous_echantillon[Y]), 2)
    print(f"\nRésultat de l'ANOVA:\neta²={es}")
    
    
def anova(nutriment):
    sns.displot(data[nutriment], kde=True)
    plt.show()
    nutrianova(data, nutriment)
    
    
interact(lambda x:anova(x), x=nutrition)
```

### Fonctions non utilisées dans l'analyse

L'objectif de ces fonctions était de vérifier la normalité des données pour l'ANOVA. Cependant, aucune variable du dataset ne respecte ce critère. Heureusement, l'ANOVA est peu sensible au déviations modérée de la loi normale. De plus, le taux de faux positifs est peu affecté par le non respect de cette hypothèse, dans la mesure ou les données ne sont pas archaiques (Glass et al. 1972, Harwell et al. 1992, Lix et al. 1996).
Dans le cas présent, les distributions sont semblables à des compositions de lois normales et à des lois normales tronquées, donc l'ANOVA devrait rester relativement fiable.

```python
def normalcheck(working_df, nutriment):
    ksr = str(stats.kstest(working_df[nutriment], 'norm'))
    ksr = ksr.split(", ")
    ksr[0] = round(float(ksr[0].split("=")[1]), 2)
    ksr[1] = float(ksr[1].split("=")[1][:-1])
    print(f"Test de normalité de Kolmogorov-Smirnov:\np value = {ksr[1]}")


def check_normality(var):
    sns.displot(data[var], kde=True)
    plt.show()
    normalcheck(data, var)
    
interact(lambda x:check_normality(x), x=numeric)
```
